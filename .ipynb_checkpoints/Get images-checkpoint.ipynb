```python
import os
from tkinter import*
import pickle
import pandas as pd
from shutil import copyfile
```

```python
class Image_Meb:
    liste_types=[]
    liste_types_standard=['W-tv-0002', 'S-0001', 'S-tv-0002', 'S-tv-0001', 'W-0001', 'ME-0002', 'ME-0001', 'S-0002', 'C-0002', 'ME-tv-0002', 'C-tv-0002', 'N-0002', 'W-tv-0001', 'N-tv-0001', 'N-tv-0002', 'C-0002_lengthAnalysis_screenshot', 'ME-tv-0001', 'N-0001', 'W-0002', 'C-0001', 'C-tv-0001', 'mapTV-2', 'map45deg-2', 'mapTV-1', 'map45deg-1']
    liste_types_standard.sort()
    
    def __init__(self,adresse):
        self.nom_fichier=os.path.basename(adresse)
        self.adresse=adresse
        self.nom_wafer=self.nom_fichier[:12]
        self.wafer_number=int(self.nom_wafer[-2:])
        self.type=self.nom_fichier[23:-4]
        
        if self.nom_fichier[-4:]==".tif":
            self.category="single"            
            self.position=self.nom_fichier[23:-12]   
        elif self.nom_fichier[-4:]==".png":
            if "map" in self.nom_fichier:
                self.category="map" 
            if "lengthAnalysis" in self.nom_fichier:
                self.category="length_analysis"
        else:
            self.category="?"
            
        if "TV-" in self.nom_fichier:
            self.angle="TV"
        else:
            self.angle="45deg"
            
        if self.type not in Image_Meb.liste_types:
            Image_Meb.liste_types.append(self.type)
```

```python
#Pickler and depickler
#Adapted from https://stackoverflow.com/questions/11218477/how-can-i-use-pickle-to-save-a-dict

adresse_pickler="W:/Communs/Lot Viewer/pickler"
adresse_pickler_liste_repertoires_4=adresse_pickler+"/liste_repertoires_4.txt"
adresse_pickler_liste_adresses_completes_4=adresse_pickler+"/liste_adresses_repertoires_4.txt"

def dump_data():
    with open(adresse_pickler_liste_repertoires_4, 'wb') as handle:
        pickle.dump(liste_repertoires_4, handle, protocol=pickle.HIGHEST_PROTOCOL)
    with open(adresse_pickler_liste_adresses_completes_4, 'wb') as handle:
        pickle.dump(liste_adresses_completes_4, handle, protocol=pickle.HIGHEST_PROTOCOL)
#dump_data()

def pickle_data():
    with open(adresse_pickler_liste_repertoires_4, 'rb') as handle:
        liste_repertoires_4 = pickle.load(handle)
    with open(adresse_pickler_liste_adresses_completes_4, 'rb') as handle:
        liste_adresses_completes_4 = pickle.load(handle)
    return liste_repertoires_4,liste_adresses_completes_4
```

```python
nb_images_with_grade=0
adresses_images_with_pggrade=[]
pggrades=[]

for i in range(len(liste_noms_wafers_with_images_at_west)):
    nom_wafer = liste_noms_wafers_with_images_at_west[i]
    adresse_image = liste_adresses_images_ouest[i]
    if nom_wafer in table_pg_grade_west["Wafer name"].values and "_cnd" not in adresse_image:
        pggrade=table_pg_grade_west.loc[table_pg_grade_west["Wafer name"]==nom_wafer]["PG @ W"].iloc[0]
        pggrades.append(round(pggrade))
        adresses_images_with_pggrade.append(adresse_image)
```
